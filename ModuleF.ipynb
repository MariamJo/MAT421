**20.2 Finite Difference Approximating Derivatives**

```python
#For this example I have taken the code from the textbook and altered it in order to find the derivative of sin.

#Import Pakcages
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
%matplotlib inline


#Define function
h = 0.1
x = np.arange(0, 2*np.pi, h) 
y = np.sin(x) 

#Forward Differentiation and Exact Solution
fd = np.diff(y)/h 
x_diff = x[:-1:] 
es = np.cos(x_diff) 


#Plot the approximate and exact solutions
plt.figure(figsize = (12, 8))
plt.plot(x_diff, fd, '--', label = 'Finite difference approximation')
plt.plot(x_diff, es, label = 'Exact solution')
plt.legend()
plt.show()

#This demonstrates the maximum error from the exact and approximate solution.
max_error = max(abs(es - fd))
print("Error is: ", max_error)
```

**20.3 Approximating of Higher Order Derivatives**

```python
#For this example I did a similar thing as above and attempted to convert the textbook code into one that could work for the cosine function.
#So I tried to change the for loop to represent a taylor series for cosine. 
#The entirety of this demonstrates different orders that can be approximated using Taylor Series

#Import Paackages
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')

#Define x and y 
x = np.linspace(-np.pi, np.pi, 200)
y = np.zeros(len(x))

labels = ['First Order', 'Third Order', 'Fifth Order', 'Seventh Order']

#Uses the loop to represent the taylor series 
plt.figure(figsize = (10,8))
for i, label in zip(range(4), labels):
    y = y + ((-1)**i * (x)**(2*i)) / np.math.factorial(2*i)
    plt.plot(x,y, label = label)

#Plots the grid
plt.plot(x, np.cos(x), 'k', label = 'Analytic')
plt.grid()
plt.title('Taylor Series Approximations of Various Orders')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

**20.4 Numerical Differentiation with Noise**

```python
#Similar to before I am adabting the examples in the textbook in order to see how noise works with sine function instead.
#The first plot shows the regular sine function while the other demonstrates the derivative of sine.

#Import Packages
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
%matplotlib inline

#Initialize variables
x = np.arange(0, 2*np.pi, 0.01) 
omega = 100
epsilon = 0.01

#Defines the function with noise
y = np.sin(x) 
y_noise = y + epsilon*np.sin(omega*x)

#Plots the first graph
plt.figure(figsize = (12, 8))
plt.plot(x, y_noise, 'r-', label = 'sin(x) + noise')
plt.plot(x, y, 'b-', label = 'sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

#Define variables for second plot
x = np.arange(0, 2*np.pi, 0.01) 
# Defines the function and noise
y = np.cos(x) 
y_noise = y + epsilon*omega*np.sin(omega*x)

#Plots the second graph
plt.figure(figsize = (12, 8))
plt.plot(x, y_noise, 'r-', label = 'Derivative sin(x) + noise')
plt.plot(x, y, 'b-', label = 'Derivative of sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
