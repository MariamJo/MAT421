**3.2 Continuity and Differentiation**

```python
#For this example I have a set expression where python uses a sympy import in order to find the limit as x goes to zero.

#Import Packages
from sympy import *

#Set the Symbol
x = symbols('x')
#Given expression
expression = (3 ** x -1)/(6 ** x -1);
#Prints the original expression
print("Expression: {}".format(expression))
#This is the function that will obtain the limit as x goes to zero.
lim = limit(expression,x,0)
#Prints the limit
print("The limit as x goes to zero is : {}".format(lim))
```

```python
#For this example I have the expression set to user input and it is the same as above but with sin(x)/x

#Import Packages
from sympy import *

#Set the Symbol
x = symbols('x')
#Input expression
expression = input()
#Prints the original expression
print("Expression: {}".format(expression))
#This is the function that will obtain the limit as x goes to zero.
lim = limit(expression,x,0)
#Prints the limit
print("The limit as x goes to zero is : {}".format(lim))
```

```python
#For this example I am computing the derivative of cos(x) and graphing both the original expression and it's derivative

#Import Packages
from sympy import *
import matplotlib.pyplot as plt
import numpy as np

#Set the Symbol
x = symbols('x')

#This obtains the derivative and prints it
derivative = diff(cos(x))
print("The derivative of the expression is : {}".format(derivative))


#Function definition
def f(x):
  return np.cos(x)
#Derivative definition
def g(x):
    return -1 * np.sin(x)

#Graphs the functions together
x = np.linspace(-10, 10, 100)
plt.plot(x, f(x))
plt.plot(x, g(x))
plt.show()

```

```python
#Finally this is a simple example I found where Taylor Series can be seen.

#Import Packages
import math

#Set variables
x = 4
y = 0

#For loop utilizing initialized variables with the taylor expression
for i in range(10):
  y+=x**i/math.factorial(i)

#Prints the outputs
print("Using Taylor Series Python Functions: ", y)
print("Using python exponent function: ", math.exp(4))
```

**3.3 Unconstrained Optimization**

```python
#This code should find the global minimum of a given function within a set interval

#Import packages
import numpy as np
from scipy.optimize import basinhopping

#Function initialization
function = lambda x: np.cos(x)
#Starting Internval
x0=[1.]

#Minimizer function along with basinhoping
minimizer_kwargs = {"method": "BFGS"}
bh = basinhopping(function, x0, minimizer_kwargs=minimizer_kwargs,niter=200)

#Prints the output
print("global minimum: x = %.4f, f(x0) = %.4f" % (bh.x, bh.fun))
```

```python
#Finally this code example demonstrates the gradient descent

#Import Packages
import numpy as np
import matplotlib.pyplot as plt

#Gradient Descent definition with different parameters
def gradient_descent(grad, start, learn, iter=50, tol=1e-06):
  #For loop for iterating through the function and parameters
  vector = start
  for _ in range(iter):
         diff = -learn * grad(vector)
         if np.all(np.abs(diff) <= tol):
           break
         vector += diff
  return vector
#Calls the function with inputs
gradient_descent(grad=lambda v: 7 * v, start=21.0, learn=0.1)
```
