**21.4 Simpson's Rule**

```python
#This example is using the Simpson's Rule code to find the integral of cosine.

#Import Packages
import numpy as np

#Initialize Variables
a = -1*np.pi/2
b = np.pi/2
n = 15
h = (b - a) / (n - 1)
x = np.linspace(a, b, n)
f = np.cos(x)

#Simpson's Rule and Error
I_simp = (h/3) * (f[0] + 2*sum(f[:n-2:2]) \
            + 4*sum(f[1:n-1:2]) + f[n-1])
err_simp = 2 - I_simp

#Print Statements
print("Integral Estimation: ", I_simp)
print("Error: ", err_simp)

```

**21.5 Computing Integrals in Python**

```python
#This example is using the Trapezoid Rule code to find the integral of cosine.

#Import Packages
import numpy as np
from scipy.integrate import trapz
from scipy.integrate import quad 

#Initialize Variables
a = -1*np.pi/2
b = np.pi/2
n = 15
h = (b - a) / (n - 1)
x = np.linspace(a, b, n)
f = np.cos(x)

#Trapezoid Rule using two different methods
I_trapz = trapz(f,x)
I_trap = (h/2)*(f[0] + 2 * sum(f[1:n-1]) + f[n-1])

#Print Statements
print(I_trapz)
print(I_trap)

```

```python
#This example is using the built in functionalities of python quad to obtain the integral for cosine.

#Quad Initialization
I_quad, est_err_quad = \
          quad(np.cos, -1*np.pi/2, np.pi/2)

#Print Statements and Error
print("Integral Estimation: ", I_quad)
err_quad = 2 - I_quad
print("Error: ", est_err_quad, err_quad)
```

```python
#This uses the given example for the sine graph and is altered to attempt to graph cosine.

#Import Packages
from scipy.integrate import cumtrapz
import matplotlib.pyplot as plt

%matplotlib inline
plt.style.use('seaborn-poster')

#Initialize Functions
x = np.arange(-1*np.pi/2, np.pi/2, 0.01)
F_exact = np.sin(x)
F_approx = cumtrapz(np.cos(x), x)

#Plot the graph
plt.figure(figsize = (10,6))
plt.plot(x, F_exact)
plt.plot(x[1::], F_approx)
plt.grid()
plt.tight_layout()
plt.title('$F(x) = \int_{-pi/2}^{pi/2} cos(y) dy$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['Exact with Offset', 'Approx'])
plt.show()
```
